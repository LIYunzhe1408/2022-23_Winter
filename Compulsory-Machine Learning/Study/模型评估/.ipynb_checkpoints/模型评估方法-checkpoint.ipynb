```python
from scipy.io.arff import loadarff

mnist = loadarff("./mnist_784.arff")
mnist
```

```python
import pandas as pd
mnist_df = pd.DataFrame(mnist[0])
mnist_df
```

```python
X, y = mnist_df.values[:, :-1],mnist_df.values[:, -1] 
```

```python
print(X.shape, y.shape)
```

```python
X_train, X_test, y_train, y_test = X[:60000], X[60000:], y[:60000], y[60000:] 
```

# 为了保证数据之间的相互独立，要shuffle

```python
import numpy as np
shuffle_index = np.random.permutation(60000)
X_train, y_train = X_train[shuffle_index], y_train[shuffle_index]
```

# 交叉验证

* 建10次模型，每一份数据集分别作为一次验证集，所得结果取平均。调参，再用测试集
* training - 训练模型用； validation - 调参用； testing - 测试


```python
y_train_5 = (y_train == b'5')
y_test_5 = (y_test == b'5')
```

```python
from sklearn.linear_model import SGDClassifier
# 构造分类器
sgd_clf = SGDClassifier(max_iter=500, random_state=42)
sgd_clf.fit(X_train, y_train_5)
```

```python
sgd_clf.predict([X[59999]])
```

```python
y[59999]
```

```python
from sklearn.model_selection import cross_val_score
cross_val_score(sgd_clf, X_train, y_train_5, cv=3, scoring='accuracy')
```

# 自定义切分数据集交叉验证

```python
from sklearn.model_selection import StratifiedKFold
from sklearn.base import clone

# 定义切分器
skfolds = StratifiedKFold(n_splits=3, random_state=42, shuffle=True)
for train_index, validate_index in skfolds.split(X_train, y_train_5):
    clone_clf = clone(sgd_clf)
    X_train_folds = X_train[train_index]
    y_train_folds = y_train_5[train_index]
    X_validate_folds = X_train[validate_index]
    y_validate_folds = y_train_5[validate_index]
    
    clone_clf.fit(X_train_folds, y_train_folds)
    y_pred = clone_clf.predict(X_validate_folds)
    n_correct = sum(y_validate_folds == y_pred)
    print(n_correct / len(y_pred))
```

# 评估指标

```python
from sklearn.model_selection import cross_val_predict
y_train_pred = cross_val_predict(sgd_clf, X_train, y_train_5, cv=3)

from sklearn.metrics import confusion_matrix
confusion_matrix(y_train_5, y_train_5)
```

```python
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
precision_score(y_train_5, y_train_pred)
```

```python
recall_score(y_train_5, y_train_pred)
```

```python
from sklearn.metrics import f1_score
f1_score(y_train_5, y_train_pred)
```

# 阈值和ROC曲线

```python
y_scores = cross_val_predict(sgd_clf, X_train, y_train_5, cv=3, method="decision_function")
```

```python
from sklearn.metrics import precision_recall_curve
precisions, recalls, thresholds = precision_recall_curve(y_train_5, y_scores)
```

```python
thresholds.shape
```

```python
from sklearn.metrics import roc_curve
fpr, tpr, thresholds = roc_curve(y_train_5, y_scores)
```

```python
import matplotlib.pyplot as plt
def plot_roc_curve(fpr, tpr, label=None):
    plt.plot(fpr, tpr, linewidth=2, label=label)
    plt.plot([0,1], [0,1], 'k--')
    plt.axis([0,1, 0, 1])
    plt.xlabel('False Positive Rate', fontsize = 16)
    plt.ylabel('True Positive Rate', fontsize = 16)
plt.figure(figsize=(8,6))
plot_roc_curve(fpr, tpr)
plt.show
```

```python
from sklearn.metrics import roc_auc_score
roc_auc_score(y_train_5, y_scores)
```
